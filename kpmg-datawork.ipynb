```python
def dropD(df):
    return df.drop_duplicates()
    
def getAllInfo(df):
    
    #print(df.shape)
    
    #print(df.info())
    
    #print(df.isna().sum())
    
    df_tmp=pd.DataFrame( df.isna().sum() )
    
    return (  df_tmp[  (df_tmp.iloc[:,0]>0) ].index )

def replaceNa(df,k):
    
    if type(df[k]) not in ( 'int64' ,'float64' ):
        df[k].fillna("NO VALUE",inplace=True)
    else:
        df[k].fillna(0,inplace=True)
```

```python
df0= pd.read_excel("/kaggle/input/kpmg-dataset/KPMG_VI_New_raw_data_update_final.xlsx" ,
                       sheet_name='Transactions',header=1)
    
df0.head()
```

```python
df1=dropD(df0)
print(str(df1.shape)+"  "+str(df0.shape) )
    
    
df2=pd.Series( getAllInfo(df1) )
print(type(df2))
print(df2)
    
for k in df2:
    replaceNa(df1,k)
    
print(getAllInfo(df1))
print(df1.info())
    
```
